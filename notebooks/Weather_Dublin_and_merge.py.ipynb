```python
# Ensure the 'Date' column is in datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Calculate the yearly count
yearly_count = df1.groupby(pd.Grouper(key='Date', freq='Y'))['Count'].sum().reset_index()

# Rename the 'Date' column to 'Year'
yearly_count = yearly_count.rename(columns={'Date': 'Year'})

# Print the yearly count
print(yearly_count)
```

```python
# Convert date_only and Time Of Call columns in df1 to datetime format
df1['datetime'] = pd.to_datetime(df1['Date'].astype(str) + ' ' + df1['Time Of Call'].astype(str).str.zfill(2) + ':00:00')
```

```python
merged_df = pd.merge(df1, df2, left_on='datetime', right_on='date')
merged_df
```

```python
# get missing values percentage in each column
missing_values_percentage = merged_df.isnull().mean() * 100

print(missing_values_percentage)
```

```python
merged_df = merged_df.drop(['datetime'], axis=1)
merged_df['Count'] = merged_df['Count'].astype(int)
merged_df
```

```python
# Extract day of week as a numerical value (Monday=0, Sunday=6)
merged_df['day_of_week'] = merged_df['date'].dt.dayofweek

# Extract month as a numerical value (January=1, December=12)
merged_df['month'] = merged_df['date'].dt.month

# Determine if the date falls on a weekend (Saturday or Sunday)
merged_df['weekend'] = (merged_df['date'].dt.dayofweek >= 5).astype(int)

# Determine the season based on the month (Winter=1, Spring=2, Summer=3, Fall=4)
merged_df['season'] = ((merged_df['month'] % 12 + 3) // 3).astype(int)
merged_df
```
